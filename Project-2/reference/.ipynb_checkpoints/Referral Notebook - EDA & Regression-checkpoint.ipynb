### Determining significant columns to keep for Regression

```python
def correlation(dataframe,list_of_columns):
    dict = {}
    for i, column in enumerate(list_of_columns):
        dict[column] = []
        corr = round(dataframe[column].corr(dataframe['saleprice']),2)
        dict[column].append(corr)
    return dict
```

```python
# Calling dictionary function to create dictionaries of all correlation coefficients for each column.
num_dict = correlation(combined,numerical)
ordinal_dict = correlation(combined,ordinal)
dummy_dict = correlation(combined,dummy);
```

```python
# Ensuring they are all in dictionary format
print(type(num_dict))
print(type(ordinal_dict))
print(type(dummy_dict))
```

```python
# To filter the dictionary for correlation coefficients more than 0.5
def filterdict(dictionary, callback):
    newdict = dict()
    # Iterate over all the items in dictionary
    for (key, value) in dictionary.items():
        # Check if item satisfies the given condition then add to new dict
        if callback((key, value)):
            newdict[key] = value
    return newdict
```

```python
print(f'Numerical Type High Correlation: \n\n{filterdict(num_dict, lambda x : np.abs(x[1]) > [0.5])}')
print('\n')
print(f'Ordinal Type High Correlation: \n\n{filterdict(ordinal_dict, lambda x :  np.abs(x[1]) > [0.5])}')
print('\n')
print(f'Dummy Type High Correlation: \n\n{filterdict(dummy_dict, lambda x :  np.abs(x[1]) > [0.5])}')
```

```python
# Converting dictionaries to dataframes
num_cor = pd.DataFrame.from_dict(num_dict,orient='index')
num_cor_df=num_cor[np.abs(num_cor[0])>(0.5)]
num_cor_df.rename(columns={0:"correlation"},inplace=True)

ordinal_cor = pd.DataFrame.from_dict(ordinal_dict,orient='index')
ordinal_cor_df=ordinal_cor[np.abs(ordinal_cor[0])>(0.5)]
ordinal_cor_df.rename(columns={0:"correlation"},inplace=True)

dummy_cor = pd.DataFrame.from_dict(dummy_dict,orient='index')
dummy_cor_df=dummy_cor[np.abs(dummy_cor[0])>(0.5)]
dummy_cor_df.rename(columns={0:"correlation"},inplace=True)
```

```python
def subplot_regplot(dataframe,list_of_columns, suptitle, list_of_titles,datafigsize,yvariable):
    fig, ax = plt.subplots((len(list_of_columns)//4)+1, 4,figsize= datafigsize) 
    fig.suptitle(suptitle,fontsize=30)
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    plt.tight_layout(w_pad = 5.0,h_pad = 5.0,pad=5.0);    
    
    for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists
        sns.regplot(data=dataframe,x=dataframe[column],y=yvariable,ax=ax[i]) 
        ax[i].set_title(list_of_titles[i],fontdict={'fontsize': 12}) 
        ax[i].yaxis.set_visible(True)
        ax[i].xaxis.set_visible(True)
```

**MLR Assumptions**:  

- **Linearity**: 
    - As seen from the scatterplots of all 10 regresors, we observe a distinct linear relationship between independent variable (saleprice) and its regressor. Condition is **fulfilled**.
- **Independence:** 
    - As seen from the heatmap of error terms against regressors, we see that there are nearly 0 correlation to the regressors. Condition is **fulfilled**.
- **Normality**: 
    - As we observe from the scatterplot of residuals against regressors(Xi), we see that all of the residual's means centres around 0. 
    - Condition is **fulfilled**.
- **Equality of Variances:** 
    - In the scatterplot of regressors(Xi) against residual, we observe that most have a uniform error term across all values of Xi, centred around 0. However, we see that at large values of x, residuals diverge. This shows heteroskedasticity. Condition is **not fulfilled**.
- **Independence Part 2:** 
    - Based on the heatmap above, we can conclude that **none** of our regressors have significant correlation with each other. Condition is **fulfilled**.

#### Model Evaluation: Statistics

```python
def r2_adj(y_true, y_preds, p):
    n = len(y_true)
    y_mean = np.mean(y_true)
    numerator = np.sum(np.square(y_true - y_preds)) / (n - p - 1)
    denominator = np.sum(np.square(y_true - y_mean)) / (n - 1)
    return 1 - numerator / denominator
```

```python
# function to check MSE and R2 of the linear regression model
def model_evaluation(y_true, y_pred, p):
    mse = round(metrics.mean_squared_error(y_true, y_pred),2)
    rmse = round(np.sqrt(mse),2)
    r2 = round(metrics.r2_score(y_true, y_pred),4)
    r2a = round(r2_adj(y_true, y_pred, p),4)
    
    print('Mean squared error      = ', mse)
    print('Root mean squared error = ', rmse)
    print('R^2                     = ', r2)
    print('Adjusted R^2            = ', r2a)
```

```python
model_evaluation(combined['saleprice'], y_pred, X.shape[1])
```

```python
def intersection(list1, list2): 
    difference_1 = list(set(list1) - set(list2)) 
    difference_2 = list(set(list2) - set(list1)) 
    return list(set(difference_1) | set(difference_2)) 
```

# Evaluation of Different Models:
#### Defined a function that will easily show the performance of each model.

```python
def overall_evaluation(X,y,df,title_of_model):
    lr = LinearRegression()
    lr.fit(X, y)
    y_pred = lr.predict(X)
    df['y_pred']=y_pred
    plot = sns.lmplot(x='saleprice',y= 'y_pred',data=df)
    plot.fig.suptitle(title_of_model, size=20)
    plot.fig.subplots_adjust(top=.85)
    plt.show()
    
    print(model_evaluation(df['saleprice'], y_pred, X.shape[1]))  ;
    
    print("**************************************************TRAIN-TEST-SPLIT**************************************************")
    np.random.seed(42)
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state = 42)
    
    print("Shape of X_train: " + str(X_train.shape))
    print("Shape of y_train: " + str(y_train.shape))
    
    print("********************************************LINEAR REGRESSION EVALUATION********************************************")
    # Scaling data
    ss=StandardScaler()
    ss.fit(X_train)
    X_train_scaled = ss.transform(X_train)
    X_test_scaled = ss.transform(X_test)

    # Linear Regression on Train/Test Data:
    lr.fit(X_train_scaled, y_train)
    y_pred_test = lr.predict(X_test)
    sns.regplot(y_test, y_pred_test)
    plt.xlabel("True Values")
    plt.ylabel("Predictions")
    plt.suptitle("Linear Regression Model", size=20)
    plt.subplots_adjust(top=.9)
    plt.show()
    # Calculation of scores:
    lr_cv_scores = cross_val_score(lr,X_train_scaled,y_train,cv=5).mean()
    print("Base Model R2 (Basic LR) : " + str(round(R2_BasicLR,4)))
    print("Base Model R2 (Basic LR - CVS) : " + str(round(np.mean(cv_scores),4)))
    print('\n')
    print("Training R2 (Basic LR) : " + str(round(lr.score(X_train_scaled, y_train),4)))
    print("Testing R2 (Basic LR) : " + str(round(lr.score(X_test_scaled, y_test),4)))
    print("Cross Val Score (Basic LR) : " + str(round((lr_cv_scores),4)))
    print('\n')
    print("Difference in R2 (Overfitting) : " + str(round((lr.score(X_train_scaled, y_train)-lr.score(X_test_scaled, y_test))*100,2))+"%")
    
    print("********************************************LASSO REGRESSION EVALUATION********************************************")
    lasso_findalpha=LassoCV(alphas=np.linspace(0.1,5.0,10000)) 
    lasso_findalpha.fit(X_train_scaled,y_train)
    lasso_alpha = lasso_findalpha.alpha_       # From base model.
    
    # Lasso Regression
    
    lasso = Lasso(alpha=lasso_alpha)
    lasso.fit(X_train_scaled,y_train)
    lasso_cv_scores = cross_val_score(lasso,X_train_scaled,y_train,cv=5)
    lasso_R2_train = round(lasso.score(X_train_scaled,y_train),4)
    lasso_R2_test = round(lasso.score(X_test_scaled,y_test),4)
    
   
    # Showing Coefficients' importance:
    pd.Series(lasso.coef_,index=X.columns).sort_values().plot.bar(figsize=(15,3));
    plt.suptitle("Lasso Regression Coefficients - Overview", size=20)
    plt.subplots_adjust(top=.9)
    
    plt.show();
```

```python
def final_overview(X,y,df,title_of_model):
    lr = LinearRegression()
    lr.fit(X, y)
    y_pred = lr.predict(X)
    df['y_pred']=y_pred
    plot = sns.lmplot(x='saleprice',y= 'y_pred',data=df)
    plot.fig.suptitle(title_of_model, size=20)
    plot.fig.subplots_adjust(top=.85)
    plt.show()
    
    print(model_evaluation(df['saleprice'], y_pred, X.shape[1]))  ;
    
    print("**************************************************TRAIN-TEST-SPLIT**************************************************")
    np.random.seed(42)
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state = 42)
    
    print("Shape of X_train: " + str(X_train.shape))
    print("Shape of y_train: " + str(y_train.shape))
    
    print("********************************************LINEAR REGRESSION EVALUATION********************************************")
    # Scaling data
    ss=StandardScaler()
    ss.fit(X_train)
    X_train_scaled = ss.transform(X_train)
    X_test_scaled = ss.transform(X_test)

    # Linear Regression on Train/Test Data:
    lr.fit(X_train_scaled, y_train)
    print('\n')
    y_pred_test = lr.predict(X_test)
    sns.regplot(y_test, y_pred_test)
    plt.xlabel("True Values")
    plt.ylabel("Predictions")
    plt.suptitle("Linear Regression Model", size=20)
    plt.subplots_adjust(top=.9)
    plt.show()
    # Calculation of scores:
    lr_cv_scores = cross_val_score(lr,X_train_scaled,y_train,cv=5).mean()
    print("Base Model R2 (Basic LR) : " + str(round(R2_BasicLR,4)))
    print("Base Model R2 (Basic LR - CVS) : " + str(round(np.mean(cv_scores),4)))
    print('\n')
    print("Training R2 (Basic LR) : " + str(round(lr.score(X_train_scaled, y_train),4)))
    print("Testing R2 (Basic LR) : " + str(round(lr.score(X_test_scaled, y_test),4)))
    print("Cross Val Score (Basic LR) : " + str(round((lr_cv_scores),4)))
    print('\n')
    print("Difference in R2 (Overfitting) : " + str(round((lr.score(X_train_scaled, y_train)-lr.score(X_test_scaled, y_test))*100,2))+"%")
    print("********************************************RIDGE REGRESSION EVALUATION********************************************")
    # Finding best Alpha
    ridge_findalpha=RidgeCV(alphas=np.linspace(1,15,1000), store_cv_values=True) 
    ridge_findalpha.fit(X_train_scaled,y_train)
    ridge_alpha = round(ridge_findalpha.alpha_,2)
    print(f'Best Alpha for Ridge Model: {ridge_alpha}')
    print('\n')
    
    print("Training R2 (LR) : " + str(round(lr.score(X_train_scaled,y_train),4)))
    print("Testing R2 (LR) : " + str(round(lr.score(X_test_scaled,y_test),4)))

    print("Estimate of testing R2 (LR) is: " +str(round(lr_cv_scores.mean(),4)))
    print('\n')
    
    # Ridge Regression
    ridge = Ridge(alpha=ridge_alpha)
    ridge.fit(X_train_scaled,y_train)
    ridge_cv_scores = cross_val_score(ridge,X_train_scaled,y_train,cv=5)
    ridge_R2_train = round(ridge.score(X_train_scaled,y_train),4)
    ridge_R2_test = round(ridge.score(X_test_scaled,y_test),4)

    print("Training R2 (RR) : " + str(round(ridge.score(X_train_scaled,y_train),4)))
    print("Testing R2 (RR) : " + str(round(ridge.score(X_test_scaled,y_test),4)))

    print("Estimate of testing R2 (RR) is: " +str(round(ridge_cv_scores.mean(),4)))
    print('\n')
    lr_cv_mse_scores = -cross_val_score(lr, X_train_scaled,y_train, cv=5,scoring ='neg_mean_squared_error')
    lr_cv_mean_mse=lr_cv_mse_scores.mean()

    print("Training MSE (LR) : " + str(round(mean_squared_error(y_train,lr.predict(X_train_scaled)),2)))
    print("Testing MSE (LR) : " + str(round(mean_squared_error(y_test,lr.predict(X_test_scaled)),2)))
    print("Estimate of testing MSE (LR) is: " +str(round(lr_cv_mse_scores.mean(),2)))
    print('\n')
    ridge_cv_mse_scores = -cross_val_score(ridge,X_train_scaled,y_train,cv=5, scoring ='neg_mean_squared_error')
    print("Training MSE (RR) : " + str(round(mean_squared_error(y_train,ridge.predict(X_train_scaled)),2)))
    print("Testing MSE (RR) : " + str(round(mean_squared_error(y_test,ridge.predict(X_test_scaled)),2)))
    print("Estimate of testing MSE (RR) is: " +str(round(ridge_cv_mse_scores.mean(),2)))
    
    # Showing Coefficients' importance:
    pd.Series(ridge.coef_,index=X.columns).sort_values().plot.bar(figsize=(15,3));
    plt.suptitle("Ridge Regression Coefficients - Overview", size=20)
    plt.subplots_adjust(top=.9)
    
    plt.show()
    print("********************************************LASSO REGRESSION EVALUATION********************************************")
    lasso_findalpha=LassoCV(alphas=np.linspace(0.1,5.0,10000)) 
    lasso_findalpha.fit(X_train_scaled,y_train)
    lasso_alpha = lasso_findalpha.alpha_       # From base model.
    print(f'Best Alpha for Lasso Model: {lasso_alpha}')
    print('\n')
    
    # Lasso Regression
    print("Training R2 (LR) : " + str(round(lr.score(X_train_scaled,y_train),4)))
    print("Testing R2 (LR) : " + str(round(lr.score(X_test_scaled,y_test),4)))

    print("Estimate of testing R2 (LR) is: " +str(round(lr_cv_scores.mean(),6)))
    print('\n')

    lasso = Lasso(alpha=lasso_alpha)
    lasso.fit(X_train_scaled,y_train)
    lasso_cv_scores = cross_val_score(lasso,X_train_scaled,y_train,cv=5)
    lasso_R2_train = round(lasso.score(X_train_scaled,y_train),4)
    lasso_R2_test = round(lasso.score(X_test_scaled,y_test),4)

    print("Training R2 (Lasso) : " + str(round(lasso.score(X_train_scaled,y_train),4)))
    print("Testing R2 (Lasso) : " + str(round(lasso.score(X_test_scaled,y_test),4)))

    print("Estimate of testing R2 (Lasso) is: " +str(round(lasso_cv_scores.mean(),6)))
    print('\n')
    lr_cv_mse_scores = -cross_val_score(lr, X_train_scaled,y_train, cv=5,scoring ='neg_mean_squared_error')
    lr_cv_mean_mse=lr_cv_mse_scores.mean()

    print("Training MSE (LR) : " + str(round(mean_squared_error(y_train,lr.predict(X_train_scaled)),2)))
    print("Testing MSE (LR) : " + str(round(mean_squared_error(y_test,lr.predict(X_test_scaled)),2)))
    print("Estimate of testing MSE (LR) is: " +str(round(lr_cv_mse_scores.mean(),2)))
    print('\n')

    lasso_cv_mse_scores = -cross_val_score(lasso,X_train_scaled,y_train,cv=5, scoring ='neg_mean_squared_error')
    print("Training MSE (Lasso) : " + str(round(mean_squared_error(y_train,lasso.predict(X_train_scaled)),2)))
    print("Testing MSE (Lasso) : " + str(round(mean_squared_error(y_test,lasso.predict(X_test_scaled)),2)))
    print("Estimate of testing MSE (Lasso) is: " +str(round(lasso_cv_mse_scores.mean(),2)))
    
    # Showing Coefficients' importance:
    pd.Series(lasso.coef_,index=X.columns).sort_values().plot.bar(figsize=(15,3));
    plt.suptitle("Lasso Regression Coefficients - Overview", size=20)
    plt.subplots_adjust(top=.9)
    
    plt.show();
```
